```python
# Diabetes dataset

# 1. Summarize the dataset using info() and describe()


# 1. Summarize the dataset using info() and describe()

df = pd.read_csv(DATA_DIR / "diabetes_012_health_indicators_BRFSS2015.csv")
print("info()")
df.info()
print("\n") #puts few lines between info and describe

print("describe()")
df.describe()
```

```python
# 2. Are there any duplicate rows


duplicate_rows = df[df.duplicated()]

print(duplicate_rows)

print("\n")
if len(duplicate_rows) > 0:
    print("There are duplicate rows.")
else:
    print("There are no duplicate rows.")
```

```python
# 3. Are there any duplicate values in a given column (when this would be inappropriate)

# Track any duplicates exist in the dataset
any_duplicates = False

# Duplicate values in a given column or not
for col in df.columns:
    if df[col].duplicated().any():
        print(f"{col}: Yes, there are duplicate values in this column")  # Duplicate values
        any_duplicates = True
    else:
        print(f"{col}: No, there are no duplicate values in this column")  # No duplicate values

print("\n")
if any_duplicates:
    print("**There are duplicate values in the dataset.**")
else:
    print("**There are no duplicate values in the dataset.**")

# Duplicate values in a given column that would be inappropriate that needs to be a unique identifier does not exist in this dataset.
```

```python
# 4. What are the mean, median, and mode of each column?


# Mean, median, mode for each column
print("Mean of each column:\n", df.mean(numeric_only=True))
print("\nMedian of each column:\n", df.median(numeric_only=True))
print("\nMode of each column:\n", df.mode().iloc[0])
```

```python
# 5a. Are there any missing or null values?
# 5b. Do you want to fill in the missing value with a mean value? A value of your choice? Remove that row?


if df.isnull().values.any():
    print("Some missing/null values in the dataset.")
else:
    print("No missing/null values in the dataset.")

# Nothing to fill in because no missing/null values.
```

```python
# 6. Identify any other inconsistent data (e.g. someone seems to be taking an action before they are born)


# Physically active but reports that they are having difficulty walking (not logical)
inconsistent_activity = df[
    (df['PhysActivity'] == 1) &
    (df['DiffWalk'] == 1)
]

# In good health but shows high number of sick days (contradict each other)
inconsistent_self_rating = df[
    (df['GenHlth'] == 1) &
    ((df['PhysHlth'] >= 25) | (df['MentHlth'] >= 25))
]

# Combine the inconsistent data (drop duplicates)
inconsistent_combined = pd.concat([
    inconsistent_activity,
    inconsistent_self_rating
]).drop_duplicates()

# Summary of inconsistency
print(f"Inconsistent due to activity vs walking difficulty: {len(inconsistent_activity)} rows")
print(f"Inconsistent due to being in good health but high number of sick days: {len(inconsistent_self_rating)} rows")
print(f"Total unique inconsistent rows: {len(inconsistent_combined)}")

inconsistent_combined.head()
```

```python
# 7. Encode any categorical variable (e.g. with one-hot encoding)
# Load original data
df = pd.read_csv(DATA_DIR / "diabetes_012_health_indicators_BRFSS2015.csv")

# One-hot encode: 'Diabetes_012' column
diabetes_dummies = pd.get_dummies(df['Diabetes_012'], prefix='Diabetes')

# Add encoded columns to the original dataset (without dropping the original)
df_combined = pd.concat([df, diabetes_dummies], axis=1)

# Save the new df to a csv
df_combined.to_csv(OUTPUT_DIR / "diabetes_dataset_with_onehot.csv", index=False)

# Show few results after one-hot encoding
df_combined.head()
```

The diabetes dataset is usable for classification-style modeling. It is structured, mostly numeric, and does not show missing values in the original checks. The main modeling concern is class imbalance in the `Diabetes_012` target variable, which may require class weighting, resampling, or careful metric selection.


# 2. Life Expectancy Dataset


```python
# 1. Summarize the datasets using info() and describe()


df = pd.read_csv(DATA_DIR / "Life Expectancy Data.csv")

print("info()")
df.info()

print("\n")

print("describe()")
df.describe()
```

```python
# 2. Are there any duplicate rows?

duplicate_rows = df[df.duplicated()]

print(duplicate_rows)

print("\n")
if len(duplicate_rows) > 0:
    print("There are duplicate rows.")
else:
    print("There are no duplicate rows.")
```

```python
# 3. Are there any duplicate values in a given column (when this would be inappropriate?)
# Track any duplicates exist in the dataset
any_duplicates = False

# Duplicate values in a given column or not
for col in df.columns:
    if df[col].duplicated().any():
        print(f"{col}: Yes, there are duplicate values in this column")  # Duplicate values
        any_duplicates = True
    else:
        print(f"{col}: No, there are no duplicate values in this column")  # No duplicate values

print("\n")
if any_duplicates:
    print("**There are duplicate values in the dataset.**")
else:
    print("**There are no duplicate values in the dataset.**")

# Duplicate values in a given column that would be inappropriate that needs to be a unique identifier does not exist in this dataset.
```

Duplicate values within individual columns are expected in this dataset because many countries share the same year, status, and health indicator values. The more important duplicate check is whether the same `Country` and `Year` combination appears more than once.


```python
# 4. What are the mean, median, and mode of each column?

# Mean, median, mode for each column
print("Mean of each column:\n", df.mean(numeric_only=True))
print("\nMedian of each column:\n", df.median(numeric_only=True))
print("\nMode of each column:\n", df.mode().iloc[0])
```

```python
# 5a. Are there any missing or null values?

# 5b. Do you want to fill in the missing value with a mean value?  A value of your choice?  Remove that row?

if df.isnull().values.any():
    print("Some missing/null values in the dataset.")
else:
    print("No missing/null values in the dataset.")
```

```python
total_rows = len(df)          # same as df.shape[0]
missing_rows = df.isna().any(axis=1).sum()

print("Total rows:", total_rows)
print("Rows with at least one missing/null:", missing_rows)
```

For the Life Expectancy dataset, many rows contain at least one missing value. Instead of dropping all rows with missing predictors, the workflow keeps rows with a nonmissing target value and imputes remaining predictors using median or mode values.


```python
# If target is "Life Expectancy" column used then drop the few rows is ok.
df.columns = df.columns.str.strip()
df = df.dropna(subset=["Life expectancy"])
```

```python
# Impute remaining missing values:
# - numeric columns -> median
# - categorical columns -> mode

# fill numeric with median (exclude Year just in case)
num_cols = df.select_dtypes(include="number").columns.drop("Year", errors="ignore")
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# fill categorical with mode
cat_cols = df.select_dtypes(exclude="number").columns
for c in cat_cols:
    df[c] = df[c].fillna(df[c].mode()[0])
```

```python
print("Total missing values:", df.isna().sum().sum())
```

```python
# 6. Identify any other inconsistent data (e.g. someone seems to be taking an action before they are born.)

# Load the ORIGINAL (raw) dataset
df_original = pd.read_csv(DATA_DIR / "Life Expectancy Data.csv")
df_original.columns = df_original.columns.str.strip()  # fixes trailing spaces in column names

# Step 6: simple inconsistency checks with the original data

#Duplicate Country-Year rows
print("Duplicate (Country, Year) rows:",
      df_original.duplicated(["Country", "Year"]).sum())

#Rows with any negative numeric values
num = df_original.select_dtypes(include="number")
print("Rows with any negative numeric value:",
      (num < 0).any(axis=1).sum())

#Percent-style columns outside 0–100
pct_cols = ["Hepatitis B", "Polio", "Diphtheria"]
for c in pct_cols:
    if c in df_original.columns:
        out = ((df_original[c] < 0) | (df_original[c] > 100)).sum()
        print(f"{c} outside 0–100:", out)
```

The original Life Expectancy data did not show duplicate `Country`-`Year` rows, negative numeric values, or out-of-range values for selected vaccination percentage columns in the saved checks.


```python
# Save with preferred filename
df.to_csv(OUTPUT_DIR / "Life_Expectancy_Cleaned.csv", index=False)
print("Saved as Life_Expectancy_Cleaned.csv")
```

```python
# 7. Encode any categorical variables (e.g. with one-hot encoding.)

# Step 7: One-hot encode categorical variables
df_clean = pd.read_csv(OUTPUT_DIR / "Life_Expectancy_Cleaned.csv")
df_clean.columns = df_clean.columns.str.strip()

cat_cols = df_clean.select_dtypes(exclude="number").columns
df_encoded = pd.get_dummies(df_clean, columns=cat_cols, drop_first=True)

df_encoded.head()
```

```python
print("Rows, Columns:", df.shape)
```

```python
df_encoded.to_csv(OUTPUT_DIR / "Life_Expectancy_OneHot.csv", index=False)
print("Saved as Life_Expectancy_OneHot.csv")
```

## Life Expectancy conclusion

The Life Expectancy dataset is usable after cleaning. The key issue is missing data, so the notebook removes rows with missing target values and imputes remaining missing predictors. The dataset is appropriate for regression-style modeling because life expectancy is continuous.


# 3. SUPPORT2 Dataset


```python
# 1. Summarize the datasets using info() and describe()

df = pd.read_csv(DATA_DIR / "support2.csv")

print("info()")
df.info()

print("\n")

print("describe()")
df.describe()
```

```python
# 2. Are there any duplicate rows?

duplicate_rows = df[df.duplicated()]

print(duplicate_rows)

print("\n")
if len(duplicate_rows) > 0:
    print("There are duplicate rows.")
else:
    print("There are no duplicate rows.")
```

```python
# 3. Are there any duplicate values in a given column (when this would be inappropriate?)

# Track any duplicates exist in the dataset
any_duplicates = False

# Duplicate values in a given column or not
for col in df.columns:
    if df[col].duplicated().any():
        print(f"{col}: Yes, there are duplicate values in this column")  # Duplicate values
        any_duplicates = True
    else:
        print(f"{col}: No, there are no duplicate values in this column")  # No duplicate values

print("\n")
if any_duplicates:
    print("**There are duplicate values in the dataset.**")
else:
    print("**There are no duplicate values in the dataset.**")

# Duplicate values in a given column that would be inappropriate that needs to be a unique identifier does not exist in this dataset.
```

Duplicate values within individual SUPPORT2 columns are expected because many patients share common clinical categories, diagnoses, or outcome values. Duplicate row checks are more important than repeated values in individual columns.


```python
# 4. What are the mean, median, and mode of each column?

# Mean, median, mode for each column
print("Mean of each column:\n", df.mean(numeric_only=True))
print("\nMedian of each column:\n", df.median(numeric_only=True))
print("\nMode of each column:\n", df.mode().iloc[0])
```

```python
# 5a. Are there any missing or null values?

# 5b. Do you want to fill in the missing value with a mean value?  A value of your choice?  Remove that row?

# Example: don't impute target columns (change if actual target(s) is different later on)
target_cols = ["death"]  # adjust later if dataset uses different names

# fill numeric (except targets) with median
num_cols = df.select_dtypes(include="number").columns.difference(target_cols)
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# fill categorical with mode
cat_cols = df.select_dtypes(exclude="number").columns
for c in cat_cols:
    df[c] = df[c].fillna(df[c].mode()[0])
```

```python
print("Targets in df:", [c for c in target_cols if c in df.columns])
print("Missing in targets after:", df[target_cols].isna().sum() if all(c in df.columns for c in target_cols) else "Target col not found")
```

```python
print("Total missing values:", df.isna().sum().sum())
```

```python
# 6. Identify any other inconsistent data (e.g. someone seems to be taking an action before they are born.)

# Duplicate rows
print("Duplicate rows:", df.duplicated().sum())

# Negative values where they usually should not be negative
nonneg_cols = ["age","meanbp","wblc","hrt","resp","temp","pafi","alb","bili","crea","sod","ph",
               "glucose","bun","urine","charges","totcst","totmcst","scoma","sps","aps",
               "adlp","adls","adlsc","hday","surv2m","surv6m","dnrday"]

for c in nonneg_cols:
    if c in df.columns:
        neg = (df[c] < 0).sum()
        if neg > 0:
            print(f"{c}: negative values -> {neg}")

# Columns that should be 0/1 but contain other values
binary_cols = ["death","hospdead","dnr","diabetes","dementia","ca","sfdm2"]
for c in binary_cols:
    if c in df.columns:
        bad = (~df[c].dropna().isin([0, 1])).sum()
        print(f"{c}: values not in 0/1 -> {bad}")

# range checks (counts how many are outside range)
range_checks = {
    "age": (0, 120),
    "meanbp": (1, 200),     # 0 is suspicious for blood pressure
    "temp": (30, 45),
    "ph": (6.8, 7.8),
    "alb": (0, 10),         # >10 is unusual (flags very high values)
    "bili": (0, 50),        # flags extreme bilirubin
    "glucose": (0, 1000),   # flags extreme glucose
    "bun": (0, 200),        # flags extreme BUN
    "urine": (0, 5000)      # flags extreme urine output
}

for c, (lo, hi) in range_checks.items():
    if c in df.columns:
        out = ((df[c] < lo) | (df[c] > hi)).sum()
        print(f"{c}: outside {lo}-{hi} -> {out}")
```

The SUPPORT2 dataset is usable but requires cleaning. Some variables contain missing values and some clinical measurements need range checks before modeling. These checks help identify values that should be reviewed, rather than automatically removed.


```python
df.to_csv(OUTPUT_DIR / "support2_cleaned.csv", index=False)
print("Saved as support2_cleaned.csv")
```

```python
# 7. Encode any categorical variables (e.g. with one-hot encoding.)

# Step 7: One-hot encode categorical variables and save

#clean column names
df.columns = df.columns.str.strip()

# one-hot encode all non-numeric columns
cat_cols = df.select_dtypes(exclude="number").columns
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# save the encoded dataset
df_encoded.to_csv(OUTPUT_DIR / "support2_onehot.csv", index=False)
print("Saved as support2_onehot.csv")
```
